```python
import pandas as pd
```

```python
df1=pd.DataFrame({"ID":[1,2,3,4],"Name":["A","B","C","D"],"Class":[10,11,12,13]})
df1
```

```python
df2=pd.DataFrame({"ID":[3,4],"Name":["C","D"],"Class":[12,13]})
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],axis=1)
```

```python
pd.concat([df1,df2],axis=1,join="inner")
```

```python
pd.concat([df1,df2],axis=1,join="outer")
```

```python
pd.concat([df1,df2],axis=1,keys=["df1","df2"])
```

```python
pd.concat([df1,df2],axis=0,keys=["df1","df2"])
```

```python
df2=pd.DataFrame({"rollno":[11,12,13,14]})
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],sort=True)
```
