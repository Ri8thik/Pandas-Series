```python
import pandas as pd
```

```python
df1=pd.DataFrame({"A":[1,2,3,4],"B":[5,6,7,8]})
df1
```

```python
df2=pd.DataFrame({"C":[11,21,31,41],"D":[50,60,70,80]})
df2
```

```python
df1.join(df2)
```

```python
df2.join(df1)
```

```python
df2=pd.DataFrame({"C":[11,21],"D":[50,60]})
df2
```

```python
df1.join(df2)
```

```python
df1.join(df2,how="left")
```

```python
df1.join(df2,how="right")
```

```python
df1.join(df2,how="inner")
```

```python
df1.join(df2,how="outer")
```

```python
df2=pd.DataFrame({"A":[1,2,3,4],"D":[50,60,70,80]})
df2
```

```python
df1.join(df2,lsuffix="_A")
```

```python
df1.join(df2,rsuffix="_A")
```
